```python
import networkx as nx
from utils import load_graph, rodger_graphs, weighted_graphs, unweighted_graphs
```

```python
#bipartite
path = "data/networkrepository/miscellaneous/GD96_b/GD96_b.mtx"
G = load_graph(path, False)
print("is bipartite:", nx.is_bipartite(G))
N = G.number_of_nodes()
M = G.number_of_edges()
print(f"N:{N}, M:{M}")
options = {
    'node_size': 50,
}
nx.draw_random(G, **options)
```
